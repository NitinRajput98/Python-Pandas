# 8. Turn the item price into a float

```python
dollar = lambda x : float(x[1:-1])
df['item_price']=df['item_price'].apply(dollar)
```

```python
df.dtypes['item_price']
```

# 9. How much was the revenue for the period in the dataset?

```python
revenue = (df['quantity']*df['item_price']).sum()
print('The total revenue was : $ {}'.format(revenue))
```

# 11. delete the duplicates in item_name and quantity

```python
filtered = df.drop_duplicates(['item_name','quantity'])
```

# 12. select only the products with quantity equals to 1

```python
one_prod = filtered[filtered['quantity']==1]
```

# 13. select only the item_name and item_price columns

```python
price_per_item = one_prod[['item_name','item_price']]
```

# 14. sort the values from the most to less expensive

```python
price_per_item.sort_values(by = 'item_price',ascending=False)
```
